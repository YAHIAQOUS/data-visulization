# Netflix

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./csv/netflix_titles.csv')
df.tail(10)
```

## Grouping by Votes

```python
sns.set()
votes = df[['title','rating']].groupby('rating').count()
votes
```

## Realeased Years

```python
years = df[['title','release_year']].groupby('release_year').count().sort_values('title',ascending=False)
years

test = sns.relplot(x='release_year',y='title',data=years)
test
```

# Top countries

```python
years = df[['country','title']].groupby('country').count().sort_values('title',ascending=False)
years

test = sns.relplot(x='title',y='country',data=years.head(10))
test
```
