# TMDB Movies

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./csv/tmdb_movies.csv')
df.head()
```

## Top Votes

```python
sns.set()
top_votes = df[['title','vote_average']].sort_values('vote_average',ascending=False)
test = sns.barplot(x="vote_average",y="title",data=top_votes.head(10))
top_votes.head() , test
```

## Top Revenue

```python
top_revenue = df[['revenue','title']].sort_values('revenue',ascending=False)
test = sns.barplot(x="revenue",y="title",data=top_revenue.head(10))
top_votes.head() , test
```

## Top Revenue by Original Languages

```python
language = df[['original_language','revenue']].groupby('original_language').max().sort_values('revenue',ascending=False)
sns.relplot(x="original_language",y="revenue",data=language.head(10))
```
